# GRAPHS

```python
# 1) Distribution of Order Priorities
region_sales = s.groupby("Region")[["Total Revenue", "Total Profit"]].sum().sort_values("Total Revenue", ascending=False)
region_sales.plot(kind="bar", figsize=(10,6))
mtl.xticks(rotation=30)
```

Each bar represents a Region (Asia, Europe, etc.).

The bar height tells us the total revenue and total profit from that region.

Example: If Sub-Saharan Africa has a very tall bar in Revenue but shorter in Profit â†’ It means sales are high but profit margins are smaller.

```python
# 2 
c_r=s.groupby("Country")["Total Revenue"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=c_r,y=c_r.index)
```

A horizontal bar chart of Top 10 countries.

The longest bar = country with highest revenue.

Example: If India is at the top, it means India contributes the most money from sales.

```python
# 3 Online vs Offline Sales
channel_perf = s.groupby("Sales Channel")[["Total Revenue", "Total Profit"]].sum()
channel_perf.plot(kind="bar")
mtl.xticks(rotation=0)
```

A simple bar chart comparing Online vs Offline.

Two bars for each channel: one for Revenue and one for Profit.

Example: If Offline has higher Revenue but Online has higher Profit, it means online sales are more cost-effective.

```python
# 4 Order Priority Analysis
q=s.groupby("Order Priority")[["Total Revenue","Total Profit"]].sum()
q.plot(kind="bar")
mtl.xticks(rotation=0)
```

Bar chart for each Order Priority (L, M, H, C â†’ Low, Medium, High, Critical).

Revenue and Profit compared for each priority.

Example: If High priority orders give the most revenue, the company should process them faster.

```python
# 5 Shipping Days Distribution
s["Shipping Days"] = (s["Ship Date"] - s["Order Date"]).dt.days
sns.histplot(s["Shipping Days"], bins=30, kde=True)
```

A histogram (distribution) of shipping days.

The peak of the graph = the most common shipping time.

Example: If the highest peak is around 30 days, then most orders are delivered in ~1 month.
